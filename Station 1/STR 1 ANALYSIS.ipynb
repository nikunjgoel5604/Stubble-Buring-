```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the first dataset
file_path1 = 'STN1.csv'
df1 = pd.read_csv(file_path1)

# 1. Remove NaN values
df1_cleaned = df1.dropna()

# 2. Select only numeric columns for mean and mode calculation
numeric_df1 = df1_cleaned.select_dtypes(include='number')

# 3. Calculate the mean value for each numeric column
mean_values_df1 = numeric_df1.mean()

# 4. Calculate the mode value for each numeric column
mode_values_df1 = numeric_df1.mode().iloc[0]

# Display the mean and mode values
print("Mean values for Dataset 1:\n", mean_values_df1)
print("\nMode values for Dataset 1:\n", mode_values_df1)

# 5. Simple plotting method - Mean vs Mode for the dataset
plt.figure(figsize=(12, 6))

# Plot Mean and Mode values
plt.plot(mean_values_df1, label='Mean', marker='o')
plt.plot(mode_values_df1, label='Mode', marker='x')
plt.title('Mean vs Mode for Numeric Columns in Dataset 1')
plt.legend()
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# Load the dataset
df = pd.read_csv("STN1.csv")

# Print the column names to check for their existence
print("Columns in DataFrame:", df.columns)

# Check for missing data in the existing columns
print(df.isna().sum())

# List of columns that you want to include, adjusted based on actual columns present
columns_to_check = ['DATE', 'FIRECOUNT', 'WD', 'RH', 'WS', 'AT', 'GSDP', 'GSDP_CAP', 'HDI', 'PM10', 'PM25']

# Drop rows with missing values only for the columns that exist
columns_to_check = [col for col in columns_to_check if col in df.columns]
df_cleaned = df.dropna(subset=columns_to_check)

# Verify that all columns now have the same length
print(df_cleaned.isna().sum())

# Plot the data using Plotly Express
fig = px.line(
    df_cleaned, 
    x='DATE', 
    y=[col for col in columns_to_check if col != 'DATE'],  # Exclude 'DATE' from the y-values
    title='Environmental and Economic Data with Range Slider'
)
fig.update_xaxes(rangeslider_visible=True)
fig.show()
# Print the column names to check for their existence
print("Columns in DataFrame:", df.columns)
```
